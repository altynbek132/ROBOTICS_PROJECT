```python
perfs_dir = "perfs"

perfs = []
for file_name in list(os.listdir(perfs_dir)):
    perf = pd.read_table(f"{perfs_dir}/{file_name}", sep=" ")
    perf['name'] = file_name.split('.')[0]
    perfs.append(perf)
```

```python
perfs[1]
```

```python
for perf in perfs:
    plt.plot(perf['epoch'], perf['r_square'])
    plt.title('hidden layers: ' + perf['name'][0])
    plt.xlabel('epoch')
    plt.ylabel('r_square')
    plt.show()
```
